#### 조건으로 검색

- 가격이 10000 초과 상품 출력
- 가격이 10000에서 20000 사이 상품 출력

```python
# 가격이 10000 초과 상품 출력
df[df['가격']>10000]

# 가격이 10000 에서 20000 사이 상품 출력

df[(df['가격']>10000)&(df['가격']<20000)]
```
