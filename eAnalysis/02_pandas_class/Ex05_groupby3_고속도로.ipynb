## groupby()


#### groupby()와 함께 사용하는 집계 메소드

- count : 누락값을 제외한 데이타 수를 반환
- size : 누락값을 포함한 데이타 수를 반환
- sum
- mean
- std
- var
- min
- max
- quantile(q=0.25)
- quantile(q=0.50)
- quantile(q=0.75)
- describe
- first
- last
- nth   : n번째 행 반환
- sem   : 평균의 표준편차 반환

```python
import pandas as pd

df = pd.read_excel('data/고속버스.xlsx') 
df
```

```python
# 그룹화 - '선별' 단위로 각각의 노선에 대한 총 갯수
# dd = df.groupby('선별').count()
dd = df.groupby('선별')
# dd['선별'].count()
dd['선별'].count().sort_values(ascending=False)
```

```python
%matplotlib inline

# 한글처리
from matplotlib import rc
rc('font', family='Malgun Gothic')
```

```python
# 노선별 수에 대한 그래프
dd['선별'].count().plot(kind = 'bar')
```

```python
# 차종별, 노선별별 그룹화 작업 + 그래프 
# car = df.groupby('차종')

# car['차종'].count().sort_values(ascending=False)

temp = df.groupby(['차종','선별'])['총이용인원'].count()

temp

temp.plot(kind='bar')


# temp['차종','선별'].count().plot(kind='bar')
```

```python
# 총이용인원에 대한 평균
```

```python
# 교차테이블(cross tab) - 엑셀의 Pivot 유사
result = pd.crosstab(df['차종'],df['선별'])

result2 = pd.crosstab(df['차종'],df['선별'], margins=True)
result2
```

```python
result.plot(kind='bar')
```

```python
# '고속'에서 선별을 그래프 표현 (원그래프(pie))
result.loc['고속'].plot(kind='pie')
```
