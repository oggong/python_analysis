```python
# 기본적인 그래프
%matplotlib inline
import matplotlib.pyplot as plt

data = [10,15,19,30,20]

plt.plot(data)
```

```python
# 1- 그래프 틀(영역)을 만들기

plt.figure() # figure() 기본 틀 만들기
plt.plot(data)
plt.show() # 그것을 보여주시오 

# 2- 그래프 그리기
# 3- 그래프 보여주기


```

```python
# x축과 y축의 라벨 추가
plt.figure(figsize=(12,8))
plt.plot(data)
plt.title('My Sample',fontsize=22)
plt.xlabel('data')
plt.ylabel('value')
plt.grid()
plt.show()
```

1- 그래프 틀(영역)을 만들기
   > plt.figure()       
    
2- 그래프 그리기
   > plt.plot(data)     
    
3- 그래프 보여주기
   > plt.show()         

 
 ### *그래프 출력시 객체표시 같은 거 안 출력하게 하려면 뒤에 세미콜론(;) 붙이면 된다*

```python
# y축 데이타에 x축 데이타를 지정

step = ['one','two','three','four','five']
plt.plot(step,data) # x,y
plt.grid()
plt.show()

```

```python
# 속성 : linestyle , color , marker
plt.plot(step,data, linestyle='dashed', color='red', marker='o')
plt.grid()
```

```python
# 속성 축약
plt.plot(step,data,'ro--')
```
