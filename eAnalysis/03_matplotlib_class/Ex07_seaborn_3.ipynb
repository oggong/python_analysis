추가 그래프 확인

```python
import seaborn as sns                
import matplotlib.pyplot as plt  

tips = sns.load_dataset('tips')
```

(1) 시간대별로 지불금액의 평균을 나타내는 바이올린 그래프에 성별을 추가한다면

```python
plt.subplot()
sns.violinplot(x=tips['time'], y=tips['total_bill'], hue=tips['sex'])
```

(2) 지불금액과 팁에 대한 산점도 그래프에 성별을 추가한다면 

```python
# sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False, hue='sex', markers=['o', 'v'])
sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False, hue='sex', scatter_kws={'s':tips['size'] * 20})
```

FacetGrid()를 이용한 그룹별 그래프 : 문자열 분석에 적합

```python
# 시간대별로 지불 금액을 표현한 그래프
face = sns.FacetGrid(tips, col='time')
face.map(sns.distplot, 'total_bill')
```

```python
# 요일별로 지불 금액을 표현한 그래프에 성별 구분 추가
facet = sns.FacetGrid(tips, col='day', hue='sex')
facet.map(sns.barplot, 'sex', 'total_bill')
facet.add_legend()
```

```python
# 시간대별 흡연여부별 지불 금액을 표현한 그래프에 성별 구분 추가
facet = sns.FacetGrid(tips, col='smoker', row='time', hue='sex')
facet.map(sns.barplot, 'sex', 'total_bill')
facet.add_legend()
```
