```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글깨짐처리
font_loc = "c:/Windows/Fonts/malgun.ttf"
font_name=font_manager.FontProperties(fname=font_loc).get_name()
rc('font', family=font_name)

# 데이타 읽어오기
data=pd.read_csv('dataset/accidentdata.csv')
```

```python
# (1) 교차분석으로 요일별 발생지시도별 데이터 건수 확인 (교차분석-> crosstab )
df_pv=pd.crosstab(data['요일'],data['발생지시도'])
df_pv
```

```python
# (2) 각 행의 합이 1이 되도록 비율로 설정
# df_pv.sum() # 지역별(열단위) 합계
# df_pv.sum(1) # 각요일별(행단위) 값의 합계
# df_pv.div( df_pv.sum(1).astype(float), axis=0 ) # 전체 합계를 1로 가정할 때의 소숫점 결과
result = df_pv.div( df_pv.sum(1).astype(float), axis=0 ) 
result
```

```python
# (3) 누적막대 그래프로 시각화
result.plot(kind='bar', stacked=True)  # stacked=True : 누적막대
```

```python
# 발생시시도별 교통사고 요일분석
df_pv2=pd.crosstab(data['발생지시도'], data['요일'])
result2 = df_pv2.div( df_pv2.sum(1).astype(float), axis=0 ) 
result2.plot(kind='bar', stacked=True)  
```
