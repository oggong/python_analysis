## [예] 경기도에서 교통사망사고가 많은 5지역을 분석하여 도식화

(1) 발생지시도가 경기도 데이타셋 필터링

(2) 경기도내에서도 발생지시군구의 그룹핑

(3) 사망자수의 합계로 차트 만들기


```python

%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글깨짐처리
font_loc = "c:/Windows/Fonts/malgun.ttf"
font_name=font_manager.FontProperties(fname=font_loc).get_name()
rc('font', family=font_name)
```

```python
# 데이타 읽어오기
data=pd.read_csv('data/accidentdata.csv')

# (1) 발생지시도가 경기도인 필터링
df1=data[data.발생지시도=='경기']
df1  # 경기도인지 확인 :df1['발생지시도']
```

```python
# (2) 경기도내에서도 발생지시군구의 그룹핑
df2 = df1.groupby('발생지시군구')
df2 # 
```

```python
# (3) 사망자수의 합계로 차트 만들기
df2['사망자수'].sum()
```

```python
df3 = df2['사망자수'].sum()
# ?df3 : df3 의 데이타 타입 확인 (-> Series)
df3.plot()
```

```python
# 경기도 내에서 사망자 순 5개 지역 추출
# df3.sort_values()
# df3.sort_values(ascending=False)
# df3.sort_values(ascending=False).head(5)
# df3.sort_values().tail(5)
result = df3.sort_values(ascending=False).head(5)
result
# ?result에 shift+tab으로 Series 자료형 확인
```

```python
# 결과 차트로 시각화 ( pandas 차트)
result.plot(kind='pie', title='2012~14 경기도내 교통사망사고')
```

```python
# matplotlib.pyplot 이용
col = ['red','orange','yellow','green','blue']
plt.pie(result, colors=col, autopct='%.2f') # autopct='%.2f': 소숫점표시
plt.title('2012~14 경기도내 교통사망사고 빈발도시')
```
