** 양적자료로 빈도와 백분율을 구하려면 구간별 새로운 자료를 생성하고 구함

### 양적자료 시각화 

1. 히스토그램

2. 박스플롯



```python
# 데이타 읽어오기

import pandas as pd

# 인덱스가 2번 생성되기에 기존 인덱스를 인덱스로 지정
diamonds = pd.read_csv('data/diamonds.csv', index_col='Unnamed: 0')
```

```python
diamonds
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc

plt.rcParams['axes.unicode_minus']=False  # 추가설정 : 폰트를 변경하면 -표시가 ㅁ으로 변경되기에 '-'를 변경하지 않도록 지정
rc('font', family='Malgun Gothic')
```

```python
# price 에 대해서 히스토그램
axes = plt.subplots()
axes = sns.distplot(diamonds['price'])
axes.set_title('diamonds price')
```

```python
# price 에 대해서 박스 플롯
```

### [양적자료를 질적자료로 새로 생성 ]

```python
diamonds['gprice'] = pd.cut(diamonds['price'],[0,5000,10000,15000,20000],labels=['5000','10000','15000','20000'])
diamonds.head(20)
```

```python
# gprice -> 막대 그래프 그릴수 있다우
a = diamonds.groupby('gprice').count()['price']
plt.subplot()
axes = sns.barplot(x=a.index,y=a.values) 
```
